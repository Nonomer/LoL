```python
import pandas as pd
import numpy as np
!pip install matplotlib
import matplotlib as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier

import warnings
warnings.filterwarnings('ignore')
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 200)
```

```python
data = pd.read_csv('LeagueofLegends.csv')
```

```python
data.head()
for col in data.columns: 
    print(col)
    
df = data[['bResult','rResult','blueTopChamp','blueJungleChamp','blueMiddleChamp','blueADCChamp','blueSupportChamp','redTopChamp','redJungleChamp',
         'redMiddleChamp','redADCChamp','redSupportChamp']].copy()
```

```python
# Convert data to numeric
df.head()
bluetop,btun = pd.factorize(df['blueTopChamp'])
bluejungle,bjun = pd.factorize(df['blueJungleChamp'])
bluemid,bmun = pd.factorize(df['blueMiddleChamp'])
blueadc,baun = pd.factorize(df['blueADCChamp'])
bluesup,bsun = pd.factorize(df['blueSupportChamp'])
redtop,rtun = pd.factorize(df['redTopChamp'])
redjungle,rjun = pd.factorize(df['redJungleChamp'])
redmid,rmun = pd.factorize(df['redMiddleChamp'])
redadc,raun = pd.factorize(df['redADCChamp'])
redsup,rsun = pd.factorize(df['redSupportChamp'])
list1 = [bluetop,bluejungle,bluemid,blueadc,bluesup,redtop,redjungle,redsup,redadc,redmid]

bluetop = bluetop.tolist()
bluejungle = bluejungle.tolist()
bluemid = bluemid.tolist()
blueadc = blueadc.tolist()
bluesup = bluesup.tolist()
redtop = redtop.tolist()
redjungle = redjungle.tolist()
redmid = redmid.tolist()
redadc = redadc.tolist()
redsup = redsup.tolist()

lol_df = pd.DataFrame(np.column_stack([df['bResult'],bluetop,bluejungle,bluemid,blueadc,bluesup,redtop,redjungle,
                     redmid,redadc,redsup]), columns = ['win','bluetop','bluejungle','bluemid',
                     'blueadc','bluesup','redtop','redjungle','redmid','redadc','redsup'])

lol_df
```

```python
lol_df.corr()
features = lol_df.iloc[:, lol_df.columns != 'win']
target = lol_df.iloc[:, lol_df.columns == 'win']
forest = ExtraTreesClassifier(n_estimators = 250, random_state = 0)

forest.fit(features,target)


print(forest.feature_importances_) #use inbuilt class feature_importances of tree based classifiers


feat_importances = pd.Series(forest.feature_importances_, index=features.columns)
print(feat_importances)
```
